```python
# working code
import numpy as np

def gramschmidt(A):
    n, m = A.shape # We get the shape of the matrix A

    Q = np.zeros((n, n)) # initializing matrix Q with Zeros
    u = np.zeros((n, n)) # initializing matrix u with zeros

    u[:, 0] = A[:, 0]
    Q[:, 0] = u[:, 0] / np.linalg.norm(u[:, 0]) #numpy linear algebra used to normalize u vector

    for i in range(1, n):

        u[:, i] = A[:, i]
        for j in range(i):
            u[:, i] -= (A[:, i] @ Q[:, j]) * Q[:, j] # get each u vector

        Q[:, i] = u[:, i] / np.linalg.norm(u[:, i]) # after we get each u vector we then normalize it for the orthogonal matrix Q
    R = np.zeros((n, m))  # Initializing R matrix with zeros
    for i in range(n):
        for j in range(i, m):
            R[i, j] = A[:, j] @ Q[:, i]

    return Q, R # returns the Q, R matrices after the factorization is done

def main():

    A = np.array([[1, 3, 4], 
              [3, 1, 2],
              [4, 2, 1]])

    for i in range(20):
        Q, R = gramschmidt(A)
        A = np.dot(R, Q)
        print(f'Iteration {i+1}:')
        print (A)
        if (i == 19):
            for i in range(len(A)):
                for j in range(len(A[0])):
                    A[i][j] = int(A[i][j] * 10 ** 4) / 10 ** 4
                    stringAL = str(A[i][j])
                    p = stringAL.find(".") + 1 + 3
                    A[i][j] = (float(stringAL[0:p]))
            print ("\nApproxiamating A20 iteration:")
            print (A)
main()
```
